```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from keras.utils import np_utils
from sklearn.model_selection import train_test_split
import numpy as np

path = "../../dane/8CPU_20RAM/3600s/few_merged.csv"
```

```python
# Załaduj dane
df = pd.read_csv(path)
```

```python
df
```

```python
df = pd.get_dummies(df, columns=['endpointUrl_methods'])
```

```python
scaler = StandardScaler()

df[['cpuUsage_stock', 'memoryUsage_stock']] = scaler.fit_transform(df[['cpuUsage_stock', 'memoryUsage_stock']])
```

```python
# Kodowanie pudełek
le = LabelEncoder()
df['test'] = le.fit_transform(df['test'])
```

```python
# Wybranie tylko tych kolumn, które są potrzebne
features = df[['test', 'timestamp'] + [col for col in df.columns if 'endpointUrl_methods' in col]]
targets = df[['test', 'cpuUsage_stock', 'memoryUsage_stock']]
```

```python
targets
```

```python
# Podział na zestawy treningowe i testowe dla każdego testu
X_train_list, y_train_list, X_test_list, y_test_list = [], [], [], []
window_size = 7000
step_size = 1000
```

```python
def split_data(features, targets, test_ratio=0.2):
    split_index = int((1 - test_ratio) * len(features))
    return features[:split_index], targets[:split_index], features[split_index:], targets[split_index:]

def create_windows(data, window_size, step_size, feature_cols, target_cols):
    X, y = [], []
    for i in range(0, len(data) - window_size - step_size + 1, step_size):
        X.append(data[feature_cols].iloc[i:i+window_size].values)
        y.append(data[target_cols].iloc[i+window_size:i+window_size+step_size].values)
    return np.array(X), np.array(y)
```

```python
for test in df['test'].unique():
    test_data_features = features[features['test'] == test]
    test_data_targets = targets[targets['test'] == test]

    train_data_features = test_data_features.iloc[:int(len(test_data_features)*0.8)]
    test_data_features = test_data_features.iloc[int(len(test_data_features)*0.8):]

    train_data_targets = test_data_targets.iloc[:int(len(test_data_targets)*0.8)]
    test_data_targets = test_data_targets.iloc[int(len(test_data_targets)*0.8):]

    feature_cols = [col for col in train_data_features.columns if col != 'test']
    target_cols = [col for col in train_data_targets.columns if col != 'test']

    print(feature_cols)
    print(target_cols)

    train_result = create_windows(train_data_features, window_size, step_size, feature_cols, target_cols)
    X_train = train_result[0]
    y_train = train_result[1]
    test_result = create_windows(test_data_features, window_size, step_size, feature_cols, target_cols)
    X_test = test_result[0]
    y_test = test_result[1]

    print(type(X_train))
    print(type(y_train))

    X_train_list.append(X_train)
    y_train_list.append(y_train)
    X_test_list.append(X_test)
    y_test_list.append(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import GRU, Dense, TimeDistributed
from keras.utils import to_categorical
```

```python
print(X_train_list[0])
```

```python
# Łączenie danych z różnych testów
X_train_combined = np.concatenate(X_train_list, axis=0)
y_train_combined = np.concatenate(y_train_list, axis=0)
```

```python
# Trenowanie modelu na połączonym zestawie danych
model = Sequential()    
model.add(GRU(150, return_sequences=True, input_shape=(step_size, X_train_combined.shape[2])))
model.add(TimeDistributed(Dense(len(targets))))
model.compile(loss='mean_squared_error', optimizer='adam')

# dodatkowe metryki:
## MSE jest często używany, ponieważ kara duże błędy bardziej niż małe, ale MAE daje nam lepsze pojęcie o rzeczywistym błędzie naszych prognoz.
# model.compile(optimizer='adam', loss='mean_squared_error', metrics=[metrics.RootMeanSquaredError(), metrics.MeanAbsoluteError()])
```

```python
model.fit(X_train_combined, y_train_combined, epochs=5, batch_size=64)
```

```python
for i in range(len(X_test_list)):
    loss = model.evaluate(X_test_list[i], y_test_list[i])
    print(f"Test {i+1}: Loss = {loss}")
```
